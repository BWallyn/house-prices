## 1.3 Dataset

```python
df_train = pd.read_csv(os.path.join(path_data, 'train.csv'), index_col='Id')
y_train = df_train["SalePrice"]
df_train.drop(columns=["SalePrice"], inplace=True)
df_train
```

# 2. Create model

## 2.1 Create features

```python
df_train = create_feats(df_train)
```

## 2.2 Train model

```python
feat_cat = df_train.select_dtypes(exclude=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']).columns
feat_cat
```

```python
feat_imp = feature_imputer()
col_transf = column_transformer()



params = {
    'max_iter': 1000,
    'max_depth': 7,
    'categorical_features': feat_cat,
}

estimator = pipe_estimator(feat_imp, col_transf, **params)
```

```python
estimator.fit(df_train, y_train)
```

```python
pred_train = estimator.predict(df_train)
```
